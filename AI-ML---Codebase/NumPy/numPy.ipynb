```python
#range
arr = np.arange(1 , 10 , 2)
print(arr)
```

```python
#linspace

arr = np.linspace(1 ,100 , 4)
print(arr , arr.shape)
```

```python
#vectorization

arr = np.array([1,2,3,4,5,6])
arr2 = np.array([1,2,4,1,5,6,])
print(arr **2)
print(arr +10)
print(arr + arr2)
```

```python
#normalize

arr = np.array([[1,2],[3,4]])
mean =(arr.mean())
std =  (arr.std())

normalize_arr = (arr- mean)/std
print(normalize_arr)

```

```python
#methametical functions

arr = np.array([1,2,3,4,5,])
print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))
print(np.argmax(arr))
print(np.mean(arr))
print(np.std(arr))
print(np.median(arr))
print(np.var(arr))
```

```python
#power fxn

arr = np.array([1,2,3,4,5,])
print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```
