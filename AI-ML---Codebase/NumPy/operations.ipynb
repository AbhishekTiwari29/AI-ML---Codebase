```python
#reshape 
arr = np.array([[1,2,3],[4,5,6]])
print(arr , arr.shape)

reshape = arr.reshape((3,2))
print(reshape ,reshape.shape)

#2d => 1d

flatended = arr.flatten()
print(flatended , flatended.shape)
```

```python
#indexing

arr = np.array([1,2,3,4,5,6])
print(arr[0])
print(arr[2])
print(arr[5])
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr[1][1])
print(arr[0][0])
```

```python
#fancy indexing

arr = np.array([1,2,3,4,5,6])

idx = [0,2,5]
print(arr[idx])
```

```python
#boolean indexing

arr = np.array([1,2,3,4,5,6])
print(arr[arr % 2 ==0 ])
print(arr[arr < 2])
```

```python
#slicing

arr = np.array([1,2,3,4,5,6])
print(arr[1:2])
print(arr[1:5])
print(arr[1:])
print(arr[:5])
print(arr[::3])
```
