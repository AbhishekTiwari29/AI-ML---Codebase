```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data

        
```

```python
data = load_data("amazon_store_data.json")
print(data)
```

```python
# clean the data 
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5 }

    cleand_data = []
    unique_users = set()
    
    for user in data:
        #clean rating
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        #handle missing vals
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        #duplication

        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleand_data.append(user)

    return cleand_data
```

```python
data =  clean_data(data)
```

```python
# get meaningful insights from data

def get_insights(data):

    #avg rating
    tot_rating = 0
    for user in data:
        tot_rating +=  float(user["rating"])

    print(f"avg : {tot_rating/len(data)}")

    #poor rating
    poor_rating = 0
    for user in data:
        if( float(user["rating"]) < 3):
            poor_rating +=1
    print(f"% total poor rating : {poor_rating/len(data)*100}%")
```

```python
get_insights(data)
```
